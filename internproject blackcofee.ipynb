```python
def extract_title_data(url_id,url):
    try:
        response = requests.get(url)
        tree = html.fromstring(response.content)
        title_element = tree.xpath('//h1[@class="entry-title"]')
        title = title_element[0].text.strip() if title_element else "No Title"
        content_elements = tree.xpath('//div[@class="td-post-content tagdiv-type"]/p')
        content = '\n'.join([p.text_content().strip() for p in content_elements]) if content_elements else "No Content"

        return title,content
    except Exception as e:
        print(f"Error extracting title data from {url_id}: {str(e)}")
        er.append(url_id)
        return None
```

```python
for index, row in inputd.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    # Extract article data
    article_data = extract_title_data(url_id,url)
    print(article_data)

    if article_data:
        output_file_path = f'{url_id}.txt'
        with open(output_file_path, 'w', encoding='utf-8') as file:
            file.write(f"{article_data}")

        print(f"Article data extracted and saved for URL_ID {url_id} at {output_file_path}.")

print("Extraction process completed.")
```

```python
def count_syllables(word):
    vowels = "aeiouy"
    exceptions = ["es", "ed"]
    count = 0
    prev_char_was_vowel = False

    for char in word.lower():
        if char in vowels:
            if not prev_char_was_vowel:
                count += 1
            prev_char_was_vowel = True
        else:
            prev_char_was_vowel = False

    for exception in exceptions:
        if word.endswith(exception):
            count -= 1

    return max(1, count)
```

```python
def calculate_scores(text_content):
    words = [word for word in word_tokenize(text_content) if word.isalnum() and word not in sw]
    word_counter = Counter(words)

    total_words = len(words)
    total_sentences = len(sent_tokenize(text_content))
    word_counts_per_sentence = [len(word_tokenize(sentence)) for sentence in sent_tokenize(text_content)]

    complex_words_count = sum(1 for word in words if count_syllables(word) >= 3)
    percentage_complex_words = (complex_words_count / total_words) * 100

    positive_score = sum(word_counter[word] for word in pos)
    negative_score = sum(word_counter[word] for word in neg)
    polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)

    average_sentence_length = total_words / total_sentences
    average_words_per_sentence = sum(word_counts_per_sentence) / len(word_counts_per_sentence)
    fog_index = 0.4 * (average_words_per_sentence + percentage_complex_words)
    average_syllables_per_word = sum(count_syllables(word) * word_counter[word] for word in words) / total_words
    personal_pronoun_count = sum(1 for word in words if re.match(r'\b(?:I|we|my|ours|us)\b', word, flags=re.IGNORECASE))
    average_word_length = sum(len(word) * word_counter[word] for word in words) / total_words

    return {
        "Positive score": positive_score,
        "Negative score": negative_score,
        "Polarity": polarity_score,
        "Subjectivity": subjectivity_score,
        "Average sentence length": average_sentence_length,
        "Percentage of complex words": percentage_complex_words,
        "Fog index": fog_index,
        "Average words per sentence": average_words_per_sentence,
        "Complex word count": complex_words_count,
        "Word count": total_words,
        "Average syllables per word": average_syllables_per_word,
        "Personal pronoun count": personal_pronoun_count,
        "Average word length": average_word_length
    }
```

```python
def count_syllables(word):
    vowels = "aeiouy"
    exceptions = ["es", "ed"]
    count = 0
    prev_char_was_vowel = False

    for char in word.lower():
        if char in vowels:
            if not prev_char_was_vowel:
                count += 1
            prev_char_was_vowel = True
        else:
            prev_char_was_vowel = False

    for exception in exceptions:
        if word.endswith(exception):
            count -= 1

    return max(1, count)
```

```python
import os
import pandas as pd

# Function to process a single file and return the row data
def process_file(file_path, url):
    with open(file_path, "r", encoding="utf-8") as file:
        text_content = file.read().lower()

    # Extract the last 15 letters of the path as URL_ID
    url_id = os.path.basename(file_path)[-19:]

    # Calculate scores for the text content
    scores = calculate_scores(text_content)

    # Return the dictionary representing the row
    return {"URL_ID": url_id, "URL": url, **scores}

# Path to the directory containing the text files
files_directory = r"C:\Users\K Sriram\Desktop\project\extracted data"

# Read URLs from Excel file
excel_file_path = r"C:\Users\K Sriram\Downloads\Output Data Structure.xlsx - Sheet1.csv"
df_urls = pd.read_csv(excel_file_path)

# List of 100 URLs
urls = df_urls['URL'].tolist()

# Create an empty list to store the processed rows
rows = []

# Process each text file sequentially
for i, file_name in enumerate(os.listdir(files_directory)):
    file_path = os.path.join(files_directory, file_name)

    # Process the file and get the row data
    row_data = process_file(file_path, urls[i])
    print(row_data)

    # Append the row data to the list
    rows.append(row_data)

    # Indicate progress after each iteration
    print(f"Processed file {i + 1}/{len(os.listdir(files_directory))}")

# Create a DataFrame from the list of dictionaries
columns = ["URL_ID"] + ["URL"] + list(calculate_scores("dummy text").keys())
results_df = pd.DataFrame(rows, columns=columns)
print(results_df)

# Save the results to a CSV file
results_df.to_csv(r"C:\Users\K Sriram\Desktop\project\results\Output Data Structure.xlsx.csv", index=False)
```
